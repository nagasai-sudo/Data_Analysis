# Project - Task 3

This notebook contains the code and pie chart for each primary category with percentage of single, Two, three-four and morethan four authors

Note - Please run this task only after running task 2

```python
import matplotlib.pyplot as plt


#Importing & reading variables and results stored in task 2
%store -r papers_df

%store -r AI_papers_df

%store -r ML_papers_df

%store -r CC_papers_df

%store -r AR_papers_df
```

The following cell calcuates no of names for each paper and creates a column. We are going to use this to calculate no of single, Two, three - four and morethan four authors papers.

Using the lambda function to count the names based on the commas. If there are 2 names there will be one comma between them. If there are 3 names, there will be 2 commas. So, We will count the comma plus one. This will give the no of names in authors coloumn.

```python

AI_papers_df['Authors_Count'] = AI_papers_df['Author'].apply(lambda x: x.count(',')+1)

ML_papers_df['Authors_Count'] = ML_papers_df['Author'].apply(lambda x: x.count(',')+1)

CC_papers_df['Authors_Count'] = CC_papers_df['Author'].apply(lambda x: x.count(',')+1)

AR_papers_df['Authors_Count'] = AR_papers_df['Author'].apply(lambda x: x.count(',')+1)
```

```python
#Checking AI_papers_df to confirm the column we created.
AI_papers_df
```

# AI Papers Pie Chart

The following cell counts the no of authors in each category.

```python
#Calculating the papers with no of authors for artificial intelligence

AI_Papers_with_1_authors = AI_papers_df.query("Authors_Count == 1")['Authors_Count'].count() # Single Authors

AI_Papers_with_2_authors =AI_papers_df.query("Authors_Count == 2")['Authors_Count'].count() # Two Authors


AI_BooleanValues = (AI_papers_df['Authors_Count'] > 2) & (AI_papers_df['Authors_Count'] < 5)

AI_Papers_with_3_to_4_authors = AI_BooleanValues.value_counts()[True] #Three to Four Authors papers

AI_Papers_with_morethan_4_authors = AI_papers_df.query("Authors_Count > 4")['Authors_Count'].count() #Morethan four authors


#Creating Pie chart for the AI papers category

AI_papers_pie = [AI_Papers_with_1_authors,
                 AI_Papers_with_2_authors,
                 AI_Papers_with_3_to_4_authors,
                 AI_Papers_with_morethan_4_authors]

Labels = ('Single Author', 'Two Authors', 'Three to Four Authors','Morethan 4 Authors') 
colors = ( '#633EBB','#BE61CA','#F2BC5E','#F13C59') 

fig1, ax1 = plt.subplots()
explode = (0, 0, 0.1, 0) 

ax1.pie(AI_papers_pie, colors = colors, explode=explode, labels=Labels,autopct='%1.1f%%',
        shadow=True)
fig1.set_size_inches(12.5, 8.5)
plt.title('Artificial Intelligence Authors Distribution')
plt.show()
```
