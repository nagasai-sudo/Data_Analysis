# Project - Task 4

This notebook contains code for creating the grouped bar graph for no of papers published in each year

```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

#importing the variables and results stored in task 2
%store -r papers_df

%store -r AI_papers_df

%store -r ML_papers_df

%store -r CC_papers_df

%store -r AR_papers_df
```

```python
papers_df
```

Extracting the year in which the paper has been published so that it becomes easy for us to plot the bar

```python
#using the pd.DatetimeIndex. year function extracts the year. And creating a coloumn in dataframe

papers_df['Upload_Year'] = pd.DatetimeIndex(papers_df['Published_date']).year

AI_papers_df['Upload_Year'] = pd.DatetimeIndex(AI_papers_df['Published_date']).year

ML_papers_df['Upload_Year'] = pd.DatetimeIndex(ML_papers_df['Published_date']).year

CC_papers_df['Upload_Year'] = pd.DatetimeIndex(CC_papers_df['Published_date']).year

AR_papers_df['Upload_Year'] = pd.DatetimeIndex(AR_papers_df['Published_date']).year
```

This particular bar is combinied for all the categories.

```python
# Calculating the papers in each year using the count function

Papers_2017 = papers_df.query("Upload_Year == 2017")['Upload_Year'].count()
Papers_2018 = papers_df.query("Upload_Year == 2018")['Upload_Year'].count()
Papers_2019 = papers_df.query("Upload_Year == 2019")['Upload_Year'].count()
Papers_2020 = papers_df.query("Upload_Year == 2020")['Upload_Year'].count()
Papers_2021 = papers_df.query("Upload_Year == 2021")['Upload_Year'].count()


#ploting the basic yearly papers count

objects = ('2017','2018','2019','2020','2021')

y_points = np.arange(len(objects))

x_points = (Papers_2017,Papers_2018,Papers_2019,Papers_2020,Papers_2021) #X axis


plt.bar(y_points, x_points, align='center', alpha=0.5)
plt.xticks(y_points, objects)
plt.ylabel('Years')
plt.title('Yearly Papers Published')
plt.show()
```

# Grouped Bar Chart for Yearly Papers Published by Category

Following cell counts the papers published in each year using the ".count" function and stores them in variable. Which will be used to plot the clustered bar chart

```python
#This section calculates Artificial Intelligence Papers yearly publication count

AI_Papers_2017 = AI_papers_df.query("Upload_Year == 2017")['Upload_Year'].count()
AI_Papers_2018 = AI_papers_df.query("Upload_Year == 2018")['Upload_Year'].count()
AI_Papers_2019 = AI_papers_df.query("Upload_Year == 2019")['Upload_Year'].count()
AI_Papers_2020 = AI_papers_df.query("Upload_Year == 2020")['Upload_Year'].count()
AI_Papers_2021 = AI_papers_df.query("Upload_Year == 2021")['Upload_Year'].count()

#This section calculates Machine Learning Papers yearly publication count
ML_Papers_2017 = ML_papers_df.query("Upload_Year == 2017")['Upload_Year'].count()
ML_Papers_2018 = ML_papers_df.query("Upload_Year == 2018")['Upload_Year'].count()
ML_Papers_2019 = ML_papers_df.query("Upload_Year == 2019")['Upload_Year'].count()
ML_Papers_2020 = ML_papers_df.query("Upload_Year == 2020")['Upload_Year'].count()
ML_Papers_2021 = ML_papers_df.query("Upload_Year == 2021")['Upload_Year'].count()


#This section calculates Computational Complexity Papers yearly publication count
CC_Papers_2017 = CC_papers_df.query("Upload_Year == 2017")['Upload_Year'].count()
CC_Papers_2018 = CC_papers_df.query("Upload_Year == 2018")['Upload_Year'].count()
CC_Papers_2019 = CC_papers_df.query("Upload_Year == 2019")['Upload_Year'].count()
CC_Papers_2020 = CC_papers_df.query("Upload_Year == 2020")['Upload_Year'].count()
CC_Papers_2021 = CC_papers_df.query("Upload_Year == 2021")['Upload_Year'].count()


#This section calculates Hardware Architecture Papers yearly publication count
AR_Papers_2017 = AR_papers_df.query("Upload_Year == 2017")['Upload_Year'].count()
AR_Papers_2018 = AR_papers_df.query("Upload_Year == 2018")['Upload_Year'].count()
AR_Papers_2019 = AR_papers_df.query("Upload_Year == 2019")['Upload_Year'].count()
AR_Papers_2020 = AR_papers_df.query("Upload_Year == 2020")['Upload_Year'].count()
AR_Papers_2021 = AR_papers_df.query("Upload_Year == 2021")['Upload_Year'].count()

```

Using the results derived in the above cell. We can plot the grouped bar chart in the following cell.

```python
#Category wise clusterd chart by year

objects = ('2017','2018','2019','2020','2021') #Year

x = np.arange(len(objects))

width = 0.20 #width of bars

#Artificial Intelligence Bars
AI_points = (AI_Papers_2017,AI_Papers_2018,AI_Papers_2019,AI_Papers_2020,AI_Papers_2021)

#Machine Learning Bars
ML_points = (ML_Papers_2017,ML_Papers_2018,ML_Papers_2019,ML_Papers_2020,ML_Papers_2021)

#Computational Complexity Bars
CC_points = (CC_Papers_2017,CC_Papers_2018,CC_Papers_2019,CC_Papers_2020,CC_Papers_2021)

#Hardware Architecture Bars
AR_points = (AR_Papers_2017,AR_Papers_2018,AR_Papers_2019,AR_Papers_2020,AR_Papers_2021)


#Actual ploting starts here
fig, ax = plt.subplots()
AI_bars = ax.bar(x-0.4,AI_points, width,color ='#F3A76E') #AI bars with customization
ML_bars = ax.bar(x-0.2,ML_points, width,color ='#A1485E') # ML bars with customization
CC_bars = ax.bar(x+0.0,CC_points, width,color = '#020082') # CC bars with cusomization
AR_bars = ax.bar(x+0.2,AR_points, width,color = 'red') # AR bars with customization

ax.set_ylabel('No of Papers')                                        #Setting y label
ax.set_xlabel('Year')                                                #setting x laebl
ax.set_title('Papers by Year for each primary category')             #Setting title
ax.set_xticks(x, objects)                                            #setting objects as x axis
ax.legend(["AI Papers", "ML Papers", "CC Papers", "AR Papers"])      #Adding a legend

#The following block of code adds paper values on top of the bars for easier understanding.
ax.bar_label(AI_bars, padding=3)                                   
ax.bar_label(ML_bars, padding=3)
ax.bar_label(CC_bars, padding=3)
ax.bar_label(AR_bars, padding=3)


fig.tight_layout()
fig.set_size_inches(12, 8)

plt.show()
```
