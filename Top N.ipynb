```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from collections import Counter

import statistics as stats
```

# Task 1

Task one involves find the top 3 authors in each category

To achieve this We are going to create sub dataframes for each of the primary category.

```python
#Artificial Intelligence dataframe
AI_papers_df = papers_df[papers_df['Primary_category'] == 'cs.AI']

#Machine Learning dataframe
ML_papers_df = papers_df[papers_df['Primary_category'] == 'cs.LG']

#Computational Complexity Dataframe
CC_papers_df = papers_df[papers_df['Primary_category'] == 'cs.CC']

#Harware Architecture Dataframe
AR_papers_df = papers_df[papers_df['Primary_category'] == 'cs.AR']
```

1) Top 3 Artificial Intelligence Authors

We are going to split the authors names first and calculate the occurence of their names in the column to determine how many papers they are involved in.

```python
#The following creates a list of individual names
AI_Paper_authors = [x for xs in AI_papers_df['Author'] for x in xs.split(',')] 

#Using the collections module to calculate names and their occurences in the above created list as dict keys.
AI_Authors = Counter(AI_Paper_authors).keys()
AI_count = Counter(AI_Paper_authors).values()

#Creating a dictionary
AI_data = {'Author': AI_Authors, 'No of Papers': AI_count}

#creating dataframe result
AI_Authors_temp = pd.DataFrame.from_dict(AI_data)

#Final dataframe that shows data in descending manner
AI_Authors_df = AI_Authors_temp.sort_values(by= 'No of Papers',ascending = False)

#Shows top 3 authors in AI Category
AI_Authors_df.head(3)
```

2) Top 3 Machine Learning authors

```python
#The following creates a list of individual names

ML_Paper_authors = [x for xs in ML_papers_df['Author'] for x in xs.split(',')]

#Using the collections module to calculate names and their occurences in the above created list as dict keys.
ML_Authors = Counter(ML_Paper_authors).keys()
count = Counter(ML_Paper_authors).values()

#Creating a dictionary
data = {'Author': ML_Authors, 'No of Papers': count}

#creating dataframe result
ML_Authors_temp = pd.DataFrame.from_dict(data)

#Final dataframe that shows data in descending manner
ML_Authors_df = ML_Authors_temp.sort_values(by= 'No of Papers',ascending = False)


#Shows top 3 authors in AI Category
ML_Authors_df.head(3)
```

3) Top 3 Computational Complexity Authors

```python
#The following creates a list of individual names
CC_Paper_authors = [x for xs in CC_papers_df['Author'] for x in xs.split(',')]

#Using the collections module to calculate names and their occurences in the above created list as dict keys.
CC_Authors = Counter(CC_Paper_authors).keys()
count = Counter(CC_Paper_authors).values()

#Creating a dictionary
data = {'Author': CC_Authors, 'No of Papers': count}

#creating dataframe result
CC_Authors_temp = pd.DataFrame.from_dict(data)

#Final dataframe that shows data in descending manner
CC_Authors_df = CC_Authors_temp.sort_values(by= 'No of Papers',ascending = False)


#Shows top 3 authors in AI Category
CC_Authors_df.head(3)
```

4) Top 3 Hardware Architecture Authors

```python
#The following creates a list of individual names

AR_Paper_authors = [x for xs in AR_papers_df['Author'] for x in xs.split(',')]

#Using the collections module to calculate names and their occurences in the above created list as dict keys.
AR_Authors = Counter(AR_Paper_authors).keys()
count = Counter(AR_Paper_authors).values()

#Creating a dictionary
data = {'Author': AR_Authors, 'No of Papers': count}

#creating dataframe result
AR_Authors_temp = pd.DataFrame.from_dict(data)

#Final dataframe that shows data in descending manner
AR_Authors_df = AR_Authors_temp.sort_values(by= 'No of Papers',ascending = False)

#Shows top 3 authors in AI Category
AR_Authors_df.head(3)
```


# Task 3

Finding th mean and median of all the words in summary for each category

To calculate the no of words, we are going to calculate the strings after spliting the delimiter

Note: Please ignore the warning after executing this step. 

```python
#This following block of code adds new column to each primary category.

#Adding new column with no of words in summary 

AI_papers_df["Number of Words"] = AI_papers_df["Summary"].apply(lambda n: len(n.split()))
ML_papers_df["Number of Words"] = ML_papers_df["Summary"].apply(lambda n: len(n.split()))
CC_papers_df["Number of Words"] = CC_papers_df["Summary"].apply(lambda n: len(n.split()))
AR_papers_df["Number of Words"] = AR_papers_df["Summary"].apply(lambda n: len(n.split()))


#Calculating mean and median for AI Papers summary

mean_of_AI_papers_summary_words = stats.mean(AI_papers_df['Number of Words'])
median_of_AI_papers_summary_words = stats.median(AI_papers_df['Number of Words'])

#mean and median for ML papers summary

mean_of_ML_papers_summary_words = stats.mean(ML_papers_df['Number of Words'])
median_of_ML_papers_summary_words = stats.median(ML_papers_df['Number of Words'])


#mean and median for CC papers summary

mean_of_CC_papers_summary_words = stats.mean(CC_papers_df['Number of Words'])
median_of_CC_papers_summary_words = stats.median(CC_papers_df['Number of Words'])


#mean and median for AR paperssummary

mean_of_AR_papers_summary_words = stats.mean(AR_papers_df['Number of Words'])
median_of_AR_papers_summary_words = stats.median(AR_papers_df['Number of Words'])
```

Creating a dataframe to present the results in more structured way.

```python

#creating numpy array with calculate means of each primary category
mean = np.array([mean_of_AI_papers_summary_words,
         mean_of_ML_papers_summary_words,
         mean_of_CC_papers_summary_words,
         mean_of_AR_papers_summary_words])


#creating numpy array with calculate medians of each primary category
median = np.array([median_of_AI_papers_summary_words,
         median_of_ML_papers_summary_words,
         median_of_CC_papers_summary_words,
         median_of_AR_papers_summary_words])

#Categories for sake of our understanding.
category = ['Artificial Intelligence','Machine Learning', 'Computational Complexity', 'Hardware Architecture']

#Converting the data in to a dataframe.
mean_and_median = pd.DataFrame(np.column_stack([category,mean,median]),
                               columns= ['Primary Category', 'Mean of Summary Words','Median of Summary Words'])
```

# Task 4

Histograms for each primary category showing the density of papers in certain bins 

```python
#Converting the pandas series column to lists first and using them as datapoints in hists
AI = AI_papers_df["Number of Words"].tolist()
ML = ML_papers_df["Number of Words"].tolist()
CC = CC_papers_df["Number of Words"].tolist()
AR = AR_papers_df["Number of Words"].tolist()


#Creating the dimensions for the plot
fig, ([(AI_hist),(ML_hist)],[(CC_hist),(AR_hist)]) = plt.subplots(nrows=2, ncols=2, figsize =(12,10) ) 

#adding the data to Artificial Intelligence Histogram
AI_hist.hist(AI,bins = 50, color = '#FFBF00')
AI_hist.set_title('Distribution of Words for Artificial Intelligence')
AI_hist.set_ylabel('Density')
AI_hist.set_xlabel('Words')

#adding the data to Machine Learning Histogram
ML_hist.hist(ML,bins = 50, color = '#6EBD6C')
ML_hist.set_title('Distribution of Words for Machine Learning')
ML_hist.set_ylabel('Density')
ML_hist.set_xlabel('Words')

#adding the data to Computational Complexity Histogram
CC_hist.hist(CC,bins = 50, color = '#FFB759')
CC_hist.set_title('Distribution of Words for Computational Complexity')
CC_hist.set_ylabel('Density')
CC_hist.set_xlabel('Words')

#adding the data to Hardware Architecture Histogram
AR_hist.hist(AR, bins = 50, color = '#4087CF')
AR_hist.set_title('Distribution of Words for Hardware Architecture')
AR_hist.set_ylabel('Density')
CC_hist.set_xlabel('Words')

fig.tight_layout()
plt.show()
```
